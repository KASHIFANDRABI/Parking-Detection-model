```python
import os
import xml.etree.ElementTree as et

def extract_text(filename):
    """Extract information including width, height, and bounding boxes from a single XML file."""
    try:
        tree = et.parse(filename)
        root = tree.getroot()

        # Extract the image name from <filename>
        image_name_tag = root.find('filename')
        image_name = image_name_tag.text if image_name_tag is not None else "Unknown"
        
        # Extract width and height from the <size> tag
        size_tag = root.find('size')
        width = size_tag.find('width').text if size_tag is not None else "Unknown"
        height = size_tag.find('height').text if size_tag is not None else "Unknown"

        # Extract objects (bounding boxes and class IDs)
        objs = root.findall('object')
        parser = []
        for obj in objs:
            name = obj.find('name').text
            bndbox = obj.find('bndbox')
            xmin = bndbox.find('xmin').text
            xmax = bndbox.find('xmax').text
            ymin = bndbox.find('ymin').text
            ymax = bndbox.find('ymax').text

            # Append image name, width, height, and bounding box data
            parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])
        
        return parser
    except Exception as e:
        print(f"Error processing {filename}: {e}")
        return None

def process_multiple_xml_files(directory):
    """Process all XML files in the given directory."""
    all_results = []

    # List all XML files in the directory
    xml_files = [f for f in os.listdir(directory) if f.endswith('.xml')]

    for xml_file in xml_files:
        full_path = os.path.join(directory, xml_file)
        annotations = extract_text(full_path)

        if annotations is not None:
            all_results.extend(annotations)  # Append all annotations
            
    return all_results

# Example usage: replace 'your_directory_path' with the path to your XML files
directory = r'C:\Object Detection\Notes\1_datapreparation\data_images'
all_data = process_multiple_xml_files(directory)

# Print the results in the desired format
for annotation in all_data:
    print(annotation)  # Each annotation should be in the format: ['image name', 'width', 'height', 'name', 'xmin', 'xmax', 'ymin', 'ymax']
```

Save Image And Labels in Text

```python
import os
from shutil import move
```

```python
train_folder = 'data_images/train'
test_folder = 'data_images/test'


os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
cols = ['Filename','id','center_x','center_y','w','h']
groupby_obj_train = train_df[cols].groupby('Filename')
groupby_obj_test = test_df[cols].groupby('Filename')
```

```python
#groupby_obj_train.get_group('132_jpg.rf.d62de3ea40353dc0980dfd8f8ecc7299.jpg').set_index('Filename').to_csv('sample.txt',index=False,header=False)
#save each image in train and test folder andrespective labels in .txt
def save_data(Filename,folder_path, group_obj):
    #move Image
    src = os.path.join('data_images',Filename)
    dst = os.path.join(folder_path,Filename)
    move(src,dst) #move img from src to dst folder

    #save the labels
    text_filename = os.path.join(folder_path,
                                 os.path.splitext(Filename)[0]+'txt')
    group_obj.get_group(Filename).set_index('Filename').to_csv(text_filename,sep=' ',index=False,header=False) #remove commas and put separate space
```

```python
#save .txt files along with images in train follder
filename_series = pd.Series(groupby_obj_train.groups.keys()) 
```

```python
filename_series.apply(save_data,args=(train_folder,groupby_obj_train))
```

```python
#for test images
filename_series = pd.Series(groupby_obj_test.groups.keys())
filename_series.apply(save_data,args=(test_folder,groupby_obj_test))
```
